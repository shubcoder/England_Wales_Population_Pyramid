The aim of this challenge is to use a json file which contains data for every single fixture of the 2014/2015 English Premier League season and find the number of goals scored over the season by a particular team.

```python
import urllib.request, json 
```

```python
with urllib.request.urlopen('https://s3.eu-west-1.amazonaws.com/hackajob-assets1.p.hackajob/challenges/football_session/football.json') as url:
	data = json.load(url)

#reading in the json dataset 
```

```python
data
#from the output below, it can be seen that the dataset is structured as a dictionary, storing data for every single fixture of the 2014/2015 premier league season 
```

```python

def goal_count(team_name):
    matchdays = len(data['rounds'])
    #this variable allows to loop through every single matchday     

    all_games = []
    #creating an empty list which still store every single match that was played over the season 

    matches_num = len(all_games)
    #this variable allows to loop through every single fixture in the season

    all_games = []
    #creating an empty list which still store every single match that was played over the season 

    for i in range(matchdays):
        #looping through each matchday 

        matchday = data['rounds'][i]
        #obtaining information for each individual mathcday 

        fixtures = matchday['matches']
        #obtaining the matches played in each mathcday 

        fix_num = len(fixtures)
        #this variable will allow to loop through the games played in each matchday 

        for i in range(fix_num):
            all_games.append(fixtures[i])
            #add each fixture to the previously created list 

    matches_num = len(all_games)
    #this variable allows to loop through every single fixture in the season 

    team_fixtures = []
    #creating an empty list which will store every single match played over the season for a particular team 
    for i in range(matches_num):
        #looping through all the fixtures

        if all_games[i]['team1']['key'] == team_name:
            team_fixtures.append(all_games[i])
        elif all_games[i]['team2']['key'] == team_name:
            team_fixtures.append(all_games[i])
        #selecting fixtures that only include the desired team, and then adding those fixtures to the previously craeted list 

    goals = 0   
    #this is the variable which will track how many goals a team scored over the course of a season 

    team_games_len = len(team_fixtures)
    #this variable allows to loop through each fixture played by a particular team over the course of a season 

    for i in range(team_games_len):
        #looping through each game played by a chosen team 
        if team_fixtures[i]['team1']['key'] == team_name:
            goals = goals + int(team_fixtures[i]['score1'])
        elif team_fixtures[i]['team2']['key'] == team_name:
            goals = goals + int(team_fixtures[i]['score2'])
        #finding the number of goals the chosen team scored in each game and adding it to their total goal count 

    return goals
```

```python
goal_count('everton')
#calling the function will return the overall goal count for any team which took part in the 2014/2015 premier league season, from the answer below, it can be seen Everton scored 48 goals
```
