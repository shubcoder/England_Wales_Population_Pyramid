```python
import sys 
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df = pd.read_csv('Popular_Spotify_Songs.csv', encoding='ISO-8859-1')
```

```python
df1 = df.drop(['track_name', 'artist(s)_name', 'key', 'mode', 'artist_count'], axis=1)
df1["streams"] = pd.to_numeric(df["streams"], errors="coerce")
df1["in_deezer_playlists"] = pd.to_numeric(df["streams"], errors="coerce")
df1["in_shazam_charts"] = pd.to_numeric(df["streams"], errors="coerce")
df1.dropna()
```
