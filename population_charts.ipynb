This notebook will document the creation of a population pyramid for England and Wales using publicly available data from the 2021 census 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
df = pd.read_excel('census2021firstresultsenglandwales1.xlsx', 'P03', skiprows=7, usecols='B,D:AO')
df

#importing the required columns of the dataset and printing it to assess how it looks 
```

```python
popdata = df.head(1)
popdata

#in this case, only the top row is needed as it contains the required data 
```

```python
popdata_transposed = popdata.T
popdata_transposed
new_header = popdata_transposed.iloc[0] 
popdata_transposed = popdata_transposed[1:] 
popdata_transposed.columns = new_header


popdata_transposed = popdata_transposed.reset_index()
popdata_transposed = popdata_transposed.rename(columns={'index': 'Group', 'England and Wales': 'Population'})

popdata_transposed = popdata_transposed.rename_axis(None, axis=1)

popdata_transposed

female_data = popdata_transposed[popdata_transposed['Group'].str.contains('Females')]
male_data = popdata_transposed[popdata_transposed['Group'].str.contains('Males')]

female_data, male_data

#transposing the row and splitting it up into two datasets (male and female)
```

```python
female_data.loc[:, 'Group'] = female_data.loc[:, 'Group'].str.replace('Females:\nAged','').str.replace('\n[note 12]',' ')
male_data.loc[:, 'Group'] = male_data.loc[:, 'Group'].str.replace('Males:\nAged','').str.replace('\n[note 12]',' ')

female_data, male_data

#cleaning up the category labels 
```

```python
plt.barh(y = female_data['Group'], width = male_data['Population'], left = -male_data['Population'], color = '#4682b4', label = 'male')
plt.barh(y = female_data['Group'], width = female_data['Population'], color = '#ee7a87', label = 'female')
plt.xlabel('Population')
plt.ylabel('Age range')
plt.legend()
plt.xlim(xmin=-2500000, xmax = 2500000)
plt.xticks(ticks=[-2000000, -1000000, 0, 1000000, 2000000],
labels=['2,000,000', '1,000,000', '0', '1,000,000', '2,000,000'])

#creating the plot, with male data in blue, and female in pink 
```

```python
df2 = pd.read_excel('pop.xlsx', skiprows=6, usecols=[0,5])
df2.rename(columns = {'Important Notes':'Year'}, inplace = True)
df2.rename(columns = {'Unnamed: 5':'Population'}, inplace = True)
df2.head()
```

```python
population = df2['Population']
year = df2['Year']
```

```python
plt.scatter(year, population, s = 20, marker = 'x')

#creating an initial plot of the population trend to estimate the best method of fitting this data 
```

```python
#the trend loosely resembles a logistic growth curve, which is commonly observed for population trends

def func(x, L, x0, k):
    return L/(1+np.exp(-k*(x-x0)))

#defining  the logistic function, L is the maximum value for population, x0 is the midpoint of the sigmoid curve, k is the logisitc growth rate (steepness of the curve)

popt, pcov = curve_fit(func, year, population, p0 = (70000000, 2010, -0.1), maxfev = 10000)

#attempting a fit with the help of an initial guess for the parameters

xx = np.linspace(1971, 2040)
plt.scatter(year, population, s = 20, marker = 'x')
plt.plot(xx, func(xx, popt[0], popt[1], popt[2]), color = '#ee7a87')
```

```python
#the logistic curve appears to be a poor fit for this data, so an alternative method must be trialled 

z = np.polyfit(year, population, 4)
f = np.poly1d(z)

#fitting a 4th degree poloynomial 

plt.scatter(year, population, s = 20, marker = 'x')
plt.plot(xx, f(xx), color = '#ee7a87')
```

```python
#it can be seen with a polynomial fit, a far better fit for the data is achieved, this fit tells us the population is expected to peak at around 2040 
#this is an acceptable finding given our current knowledge of global population trends, especially in developed countries, with birth rates expected to decline in the near future 
```
